```python
def parse_cohpcar(cohp_path, hart=True):
    data = []
    with open(cohp_path) as file:
        start = False
        for line in file:
            if start:
                if not 'No' in line:
                    data.append(line.rstrip('\n').lstrip(' ').split(' '))
            elif 'No' in line:
                start = True
    data_out = []
    for i, line in enumerate(data):
        use = []
        for j, num in enumerate(line):
            try:
                use.append(float(num))
            except:
                continue
        data_out.append(np.array(use, dtype=float))
    data_out = np.array(data_out)
    if hart:
        data_out *= (1/27.211397)
    return np.array(data_out)
```
